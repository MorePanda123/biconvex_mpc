```python
# Simple IK for 2 DOF leg
# Author : Avadesh Meduri
# Date : 12/02/2021

import numpy as np
import pinocchio as pin
import crocoddyl

from robots import TwoDofManipulator, vis_two_dof
from action_model import DifferentialFwdKinematics
```

```python
robot_model, data = TwoDofManipulator()
state = crocoddyl.StateMultibody(robot_model)

q0 = np.array([-np.pi/6.0, np.pi/4.0]).T
x0 = np.concatenate([q0, pin.utils.zero(state.nv)])
# Running and terminal action models
DT = 1e-3
T = 100
```

```python
def trajectory_tracking_task(state, fid, rcost_arr, sn, en, traj, wt, cost_name):
    """
    This function creates a end effector tracking task in the ddp
    Input:
        state : crocoddyl multibody state
        fid : the frame id of the frame that should track the trajectory
        rcost_arr : global running cost array of the ik problem
        sn : start time index (collocation number)
        en : end time index
        traj : the trajectory to be tracked (must have shape ((en - sn), 3))
        wt : weight of the cost
        cost_name : name of the cost
    """
    for i in range(sn, en):
        Mref = crocoddyl.FrameTranslation(fid, traj[i])
        goalTrackingCost = crocoddyl.CostModelFrameTranslation(state, Mref)
        rcost_arr[i].addCost(cost_name, goalTrackingCost, wt)
        
    return rcost_arr
```

```python
terminalCostModel = crocoddyl.CostModelSum(state)
rcost_arr = []
for t in range(T):
    runningCostModel = crocoddyl.CostModelSum(state)
    rcost_arr.append(runningCostModel)

traj = []
x = -1.0
for t in range(T):
    omega = np.pi/T
    x += 2/T
    traj.append(np.array([x , -1.5 + np.sin(omega*t), 0.0]))
    
rcost_arr = trajectory_tracking_task(state, robot_model.getFrameId("FOOT"), rcost_arr, 0, T, traj, 1e2, "grippervel")
```

```python
running_cost_arr = []
for t in range(T):

    xRegCost = crocoddyl.CostModelState(state)
    uRegCost = crocoddyl.CostModelControl(state)

    rcost_arr[t].addCost("stateReg", xRegCost, 1e-4)
    rcost_arr[t].addCost("ctrlReg", uRegCost, 1e-7)
    
    runningModel = crocoddyl.IntegratedActionModelEuler(
        DifferentialFwdKinematics(state, rcost_arr[t]), DT)
    running_cost_arr.append(runningModel)

# terminalCostModel.addCost("gripperPose", goalTrackingCost, 1e5)
terminalCostModel.addCost("stateReg", xRegCost, 1e-4)
terminalCostModel.addCost("ctrlReg", uRegCost, 1e-7)    

terminalModel = crocoddyl.IntegratedActionModelEuler(
    DifferentialFwdKinematics(state,terminalCostModel), 0.)
problem = crocoddyl.ShootingProblem(x0, running_cost_arr, terminalModel)
```

```python

# Creating the DDP solver for this OC problem, defining a logger
ddp = crocoddyl.SolverDDP(problem)
log = crocoddyl.CallbackLogger()
ddp.setCallbacks([log,
                  crocoddyl.CallbackVerbose(),
                  ])
# # Solving it with the DDP algorithm
ddp.solve()
```

```python
vis_two_dof(robot_model, data, np.array(ddp.xs), 1)
```

```python
xT = ddp.xs[-1]
q = xT[0:2]
v = xT[2:]
pin.forwardKinematics(robot_model, data, q, v)
pin.updateFramePlacements(robot_model, data)
print(data.oMf[robot_model.getFrameId("FOOT")])
```
